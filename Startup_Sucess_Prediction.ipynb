```python
# Importing the modules
import pandas as pd
import numpy as np
```

```python
# Reading data from the file into a variable and displaying the first few rows.
p_file = '/content/drive/MyDrive/int303-big-data-analysis/startup_train.xlsx' # File path
p_data =  pd.read_excel(p_file)
p_data.head()
```

```python
# Checking for null values in the dataset.
p_data.isnull().sum()
```

```python
# Dropping columns with more null values and displaying the updated dataframe.
p_data = p_data.drop(['Unnamed: 6','state_code.1'], axis = 1)
p_data.head()
```

```python
# Fill missing values in 'age_first_milestone_year' and 'age_last_milestone_year' columns with 0,
# and then check for any remaining missing values in the dataframe.
p_data['age_first_milestone_year'] = p_data['age_first_milestone_year'].fillna(0)
p_data['age_last_milestone_year'] = p_data['age_last_milestone_year'].fillna(0)
p_data.isnull().sum()
```

```python
# Scaling numerical features using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

normal = ['age_first_funding_year', 'relationships', 'funding_total_usd']
p_data = p_data
for var in normal:
    p_data['normal_' + var] = MinMaxScaler().fit_transform(p_data[var].values.reshape(len(p_data), 1))
```

```python
# Displaying information about the dataset to understand its structure and data types.
p_data.info()
```

```python
# Selecting numerical features and displaying info.
x_df = p_data.select_dtypes(exclude = ['object'])
x_df.info()
```

```python
# Convert status to binary labels and concatenate with original data.
sample_data = p_data

sample_data = pd.get_dummies(sample_data['status'])
sample_data = sample_data.drop(['closed'], axis = 1)
sample_data.rename(columns = {'acquired': 'label'}, inplace = True)

frames = [p_data, sample_data]
p_data = pd.concat(frames, axis = 1)
p_data.head(20)
```

```python
# Splitting data into train and test sets with selected features.
from sklearn.model_selection import train_test_split

y = p_data['label']
X = p_data[['normal_relationships', 'normal_age_first_funding_year','normal_funding_total_usd',
          'age_last_funding_year',
          'age_first_milestone_year', 'age_last_milestone_year',
          'funding_rounds', 'milestones',
          'is_CA', 'is_NY', 'is_MA', 'is_TX', 'is_otherstate',
          'is_software', 'is_web', 'is_mobile', 'is_enterprise', 'is_advertising', 'is_gamesvideo',
          'is_ecommerce', 'is_biotech', 'is_consulting','is_othercategory',
          'has_VC', 'has_angel', 'has_roundA','has_roundB', 'has_roundC', 'has_roundD',
          'avg_participants','is_top500']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Logistic Regression model training and evaluation.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score, StratifiedKFold

log_reg =  LogisticRegression(solver='newton-cg')
log_reg.fit(X_train, y_train)


print("Train Accuracy:", log_reg.score(X_train, y_train))
print("Test Accuracy:", log_reg.score(X_test, y_test))
```
